```python
bb = df_dairy.set_index('food')['group'].to_dict()
#bb
```

```python
# step matches dairy items list with the input correlation file
G = nx.from_pandas_edgelist(df_with_endo, 'food1', 'food2', edge_attr='weight')
nx.set_node_attributes(G, bb, 'group')
```

```python
# visualizing step
nw.visualize(G)
```
